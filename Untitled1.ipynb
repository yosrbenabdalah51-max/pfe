```python
import os
import mysql.connector

chunks_folder = r"C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\chunks"

mysql_config = {
    "host": "localhost",
    "user": "root",
    "password": "Yosr2003@",
    "database": "mon_projet",
    "allow_local_infile": True
}

table_name = "sales"

conn = mysql.connector.connect(**mysql_config)
cursor = conn.cursor()

for filename in sorted(os.listdir(chunks_folder)):
    if filename.endswith(".csv"):
        file_path = os.path.join(chunks_folder, filename)
        print(f"Import de {filename} ...")

        sql = f"""
        LOAD DATA LOCAL INFILE '{file_path.replace("\\", "/")}'
        INTO TABLE {table_name}
        FIELDS TERMINATED BY ','
        OPTIONALLY ENCLOSED BY '"'
        LINES TERMINATED BY '\\n'
        IGNORE 1 ROWS
        (order_id, ref_product, is_pack, first_order, free, price, quantity, depot_id, date_time);
        """

        cursor.execute(sql)
        conn.commit()
        print(f"{filename} importé avec succès.")

cursor.close()
conn.close()

print("Import terminé.")
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

# -----------------------------
# 1️⃣ Configurer la connexion
# -----------------------------
user = "root"
password = "Yosr2003"
host = "localhost"
database = "mon_projet"

engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}/{database}")

# -----------------------------
# 2️⃣ Créer les index sur sales pour accélérer les UPDATE
# -----------------------------
with engine.begin() as conn:
    conn.execute(text("CREATE INDEX IF NOT EXISTS idx_order_id ON sales(order_id)"))

# -----------------------------
# 3️⃣ Charger le CSV par chunks
# -----------------------------
csv_path = r"C:\Users\yosrb\Desktop\data\sales.csv"
chunk_size = 100_000  # ajustable selon la RAM

for i, chunk in enumerate(pd.read_csv(csv_path, usecols=['order_id','free'], chunksize=chunk_size)):
    print(f"Traitement chunk {i+1}")
    
    # Créer une table temporaire pour ce chunk
    chunk.to_sql("temp_chunk", engine, if_exists="replace", index=False)
    
    # Créer index sur temp_chunk pour accélérer le JOIN
    with engine.begin() as conn:
        conn.execute(text("CREATE INDEX IF NOT EXISTS idx_order_id_temp ON temp_chunk(order_id)"))
    
    # Mettre à jour la colonne 'free' dans sales
    with engine.begin() as conn:
        conn.execute(text("""
            UPDATE sales AS s
            INNER JOIN temp_chunk AS t ON s.order_id = t.order_id
            SET s.free = t.free
        """))
    
    print(f"Chunk {i+1} mis à jour ✅")

print("✅ Mise à jour complète de la colonne 'free'")
```
